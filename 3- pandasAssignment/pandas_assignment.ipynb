```python
grouped = merged2.groupby('id_team')
```

```python
ss = grouped.describe()
ss.id_player['count'].sum()
```

<p>The above block shows that the total number of players is equal to <code>908</code>. But this number is actually the number of players of 35 teams not 36 teams! So we realize that data of a team has not registered.</p>

```python
chart = grouped[['goals', 'goals_conceded']].sum()
chart
```

```python
list1 = teams['team_id'].values
for i in list1:
    if i not in chart.index:
        not_valid = i
        row = teams[teams['team_id'] == i]
        not_valid_index = row['team'].index
        print("name of the club that its data hasn't registered: ", end="")
        print(row['team'].values[0])
        print(f'this is team_id: {i}')
```

```python
# droping the information of Aston Villa FC team
update_teams = teams.drop(not_valid_index)

# sorting it
update_teams.sort_values('team_id', inplace=True, ignore_index=True)
update_teams
```

```python
# creating a new column for goal difference
chart['goal_difference'] = chart['goals'] - chart['goals_conceded']
```

```python
# Sorting teams based on their goals datas
output = chart.sort_values(['goal_difference', 'goals'], ascending=[False, True])
output
```

```python
# visualization
import matplotlib.pyplot as plt

team_name = update_teams['team'].values
goals_scored = chart['goals'].values
goals_conceded = chart['goals_conceded'].values
goal_difference = chart['goal_difference'].values

fig, ax = plt.subplots(3, 1, figsize=[20,30])
ax[0].bar(team_name, goals_scored, color='blue')
ax[0].tick_params(axis='x', labelrotation=90, labelcolor='black')
ax[0].set_yticks(np.arange(0, max(goals_scored)+1, 1))
ax[0].set_ylabel("goals scored")

ax[1].bar(team_name, goals_conceded, color='red')
ax[1].tick_params(axis='x', labelrotation=90, labelcolor='black')
ax[1].set_yticks(np.arange(0, max(goals_conceded)+1, 1))
ax[1].set_ylabel("goals conceded")

ax[2].bar(team_name, goal_difference, color='green')
ax[2].tick_params(axis='x', labelrotation=90, labelcolor='black')
ax[2].set_yticks(np.arange(min(goal_difference), max(goal_difference)+1, 1));
ax[2].set_ylabel("goal difference")
fig.tight_layout()
```

```python
# merging dataframes to access players' data and attempts' data simultaneously
attempts1 = pd.merge(attempts, players)
attempts1
```

```python
attempts1['accuracy'] = attempts1['attempts_on_target'] / attempts1['total_attempts']
attempts1
```

```python
attempts2 = attempts1[attempts1['accuracy'].notna()]
attempts2
```

```python
q1 = attempts2['accuracy'].quantile(0.25)
q2 = attempts2['accuracy'].quantile(0.5)
q3 = attempts2['accuracy'].quantile(0.75)
```

```python
first_quarter = attempts2[attempts2['accuracy'] <= q1]
field_position1 = first_quarter['field_position'].mode()[0]
print(field_position1)
```

```python
second_quarter = attempts2[(attempts2['accuracy'] > q1) & (attempts2['accuracy']) <= q2]
field_position2 = second_quarter['field_position'].mode()[0]
print(field_position2)
```

```python
third_quarter = attempts2[(attempts2['accuracy'] > q2) & (attempts2['accuracy']) <= q3]
field_position3 = third_quarter['field_position'].mode()[0]
print(field_position3)
```

```python
forth_quarter = attempts2[attempts2['accuracy'] > q3]
field_position4 = forth_quarter['field_position'].mode()[0]
print(field_position4)
```
