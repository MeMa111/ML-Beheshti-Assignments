<p dir='rtl'><code>np.bincount():</code> این تابع، به عنوان ورودی ارایه ای یک بعدی دریافت میکند که <code>dtype=int</code> باید باشد. در خروجی ارایه ای میدهد که تعداد فراوانی مقدار m را در ایندکس m این ارایه به نمایش میگذارد. به عنوان مثال اگر در ارایه ورودی 5 بار عدد 3 تکرار شده باشد، در ایندکس 3 ان عدد 5 قرار میگیرد.<br>مثال بیشتر:</p>

```python
import numpy as np
np.bincount(np.arange(5))
```

```python
np.bincount(np.array([0, 1, 1, 3, 2, 1, 7]))
```

<p dir='rtl'>حال در این بخش از کد، با استفاده از تابع ذکر شده، ابتدا خروجی یک چنل عکس را دریافت میکند؛ سپس لیست را بصورت نزولی به کمک<code>np.sort()</code>مرتب میکند. در نهایت طبق خواسته مسئله 5 پیکسلی که بیشترین فراوانی را دارد بدست میاورد.</p>

```python
m = np.bincount(img_flat0)
n = np.sort(m)
max_value = n[-5]
output = m - max_value
np.where(output >= 0)
```

<h2 dir='rtl'>قسمت پنجم</h2>

<p dir='rtl'>تابع <code>neighbors_mean</code>:<br>این تابع میانگین پیکسل های اطراف پیکسل های داده شده را محاسبه میکند و خروجی را بصورت یک لیست از مقادیر میانگین میدهد. در ابتدا با استفاده از <code>np.pad()</code> استفاده شده تا برای محاسبه میانگین نقاط لبه عکس به مشکل نخورد.</p>


```python
def neighbors_mean(matrix, rows, columns, color_chanel):
    pad_matrix = np.pad(matrix, (1,), 'constant', constant_values = 0)
    mean_list = []
    for i in range(len(rows)):
        Sum = np.sum(matrix[rows[i]-1:rows[i]+2, columns[i]-1: columns[i]+2, color_chanel[i]])
        mean = (Sum - 255)//8 
        mean_list.append(mean)
        
    return mean_list
        
```

<p dir='rtl'>در این قسمت با کمک <code>np.where()</code> ایندکس نقاطی که هر چنل ان برابر 255 بوده را برمیگرداند</p>

```python
rows, columns, color_chanel = np.where(img == 255)
```

<p dir='rtl'>در این قسمت از کد ابتدا تابع را صدا میزنیم و مقادیر داخل لیست را داخل تابع <code>iter()</code> میریزد.<br> در قسمت بعدی با استفاد از <code>np.where</code> مقادیر بدست امده را جایگزین میکند.</p>

```python
my_list = iter(neighbors_mean(img, rows, columns, color_chanel))
output = np.where(img == 255, next(my_list), img)
```
