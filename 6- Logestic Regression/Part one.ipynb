# Creating our model

```python
def sigmoid(Z):
    g = 1 / (1 + np.exp(-Z))
    return g
```

```python
def compute_cost(W, b, x, y):
    m = x.shape[0]
    Z = np.dot(x, W.T) + b
    epsilon = 1e-15
    g = np.clip(sigmoid(Z), epsilon, 1 - epsilon) 
    J = (y * np.log(g)) + ((1 - y) * np.log(1 - g))
    j = J.sum()
    j = j / (-m)
    return j
    
```

```python
W = np.ones(18)
b = 0
compute_cost(W, b, x_train, y_train)
```

```python
def compute_gradient(W, b, x, y):
    m = len(x)
    Z = np.dot(x, W.T) + b
    g = sigmoid(Z)
    error = g - y
    
    d_W = np.dot(x.T, error) / m
    d_b = error.sum() / m
    
    return d_W, d_b
```

```python
W = np.ones(18)
b = 0
compute_gradient(W, b, x_train, y_train)
```

```python
def gradient_descent(W_0, b_0, x, y, num_iter, alpha):
    W = W_0
    b = b_0
    
    m = len(x)
    cost_history = []
    for i in range(num_iter):
        d_W, d_b = compute_gradient(W, b, x, y)
        
        W = W - alpha * d_W
        b = b - alpha * d_b
        
        cost = compute_cost(W, b, x, y)
        cost_history.append(cost)
        
        if i % 100 == 0:
            print(f'cost at iteration {i} is {cost:.4f}')
            
    return W, b, cost_history
        
```

```python
W_0 = np.ones(18)
b_0 = 0
num_iter = 1000
alpha = 0.01
W_opt, b_opt, cost_history = gradient_descent(W_0, b_0, x_train, y_train, num_iter, alpha)
```

# Performance Evaluation and Analysis

```python
def y_pred(W, b, x, threshold):
    Z = np.dot(x, W.T) + b
    g = sigmoid(Z)

    y_hat = np.where(g >= threshold, 1, 0)
    return y_hat
```

## Confusion Matrix

```python
def compute_metrics(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    # print(TP)
    FP = np.sum((y_true == 0) & (y_pred == 1))
    # print(FP)
    FN = np.sum((y_true == 1) & (y_pred == 0))
    # print(FN)
    TN = np.sum((y_true == 0) & (y_pred == 0))
    # print(TN)
    
    accuracy = (TP + TN) / len(y_true)
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0
    
    metrics = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1_score
    }
    
    return metrics
```

```python
prediction = y_pred(W_opt, b_opt, x_train, 0.5)
metrics = compute_metrics(y_train, prediction)
for metric, value in metrics.items():
    print(f"{metric}: {value:.2f}")
```

<p dir='rtl'> معیار accuracy برای دیتاست مورد نظر گزینه مناسبی نیست؛ چراکه دیتاست نامتوازن است. بدین معنا که نمونه های کلاس ها تعداد برابری نیستند. در این حالت با وجود اینکه مدل ما از accuracy بالایی برخوردار است، قادر به پیش بینی کلاس <code>dead=1</code> نیست. برای تحلیل بهتر میتوانیم از <code>recall, F1-score</code> استفاده کنیم.</p>
<br>
<p dir='rtl'><code>reacall</code> برابر نسبت تعداد حالت هایی است که مدل موفق به پیش بینی درست<code>dead=1</code> شده است به تعداد کل حالاتی که دیتاست <code>dead=1</code> دارد.</p>
<br>
<p dir='rtl'><code>F1-score</code> معیاری ای مناسب برای تحیلیل مدلی است که دیتاست ان نامتوازن است؛ چرا که اگر هر کدام از مقادیر <code>recall</code> یا <code>precision</code> کوچک باشند، مقدار <code>F1-score</code> نیز کوچک خواهد شد.</p>

```python
def threshold_tuning(W, b, x, y, metric, step):
    threshold_range = np.arange(step, 1, step)
    my_metric = np.zeros_like(threshold_range)
    n = len(threshold_range)
    
    precision = abs(int(np.floor(np.log10(step))))

    
    for i in range(n):
        prediction = y_pred(W, b, x, threshold_range[i])
        metrics = compute_metrics(y, prediction)
        my_metric[i] = metrics[metric]
        
    index = my_metric.argmax()
    best_threshold = threshold_range[index] 
    print(f'best threshold: {best_threshold:.{precision}f}\n{metric}: {my_metric[index]:.4f}')    
    
    return best_threshold
```

```python
threshold_tuning(W_opt, b_opt, x_train, y_train, 'F1-Score', 0.01)
```

```python
threshold_tuning(W_opt, b_opt, x_train, y_train, "Recall", 0.01)
```

```python
threshold_tuning(W_opt, b_opt, x_train, y_train, 'Precision', 0.01)
```

More accurate:

```python
best_threshold = threshold_tuning(W_opt, b_opt, x_train, y_train, 'F1-Score', 0.001)
```
