```python
import cv2
import os

ref_point = []
cropping = False

def click_and_crop(event, x, y, flags, param):
    global ref_point, cropping

    if event == cv2.EVENT_LBUTTONDOWN:
        ref_point = [(x, y)]
        cropping = True
        
    elif event == cv2.EVENT_MOUSEMOVE and cropping:
        temp_image = image.copy()  
        cv2.rectangle(temp_image, ref_point[0], (x, y), (0, 255, 0), 2)
        cv2.imshow("Image", temp_image)


    elif event == cv2.EVENT_LBUTTONUP:
        ref_point.append((x, y))
        cropping = False

        cv2.rectangle(image, ref_point[0], ref_point[1], (0, 255, 0), 2)
        cv2.imshow("Image", image)

def crop_image(input_folder, output_folder):
    global image, ref_point

    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        image = cv2.imread(input_path)
        if image is None:
            print(f"Cannot read image: {input_path}")
            continue

        clone = image.copy()
        cv2.namedWindow("Image")
        cv2.setMouseCallback("Image", click_and_crop)

        print(f"Processing: {filename}")
        while True:
            cv2.imshow("Image", image)
            key = cv2.waitKey(1) & 0xFF

            if key == ord("r"):
                image = clone.copy()
                
            elif key == ord("n"):
                cv2.imwrite(output_path, image)
                break
            
            elif key == ord("z"):
                break


            elif key == ord("c"):
                if len(ref_point) == 2:
                    crop = clone[ref_point[0][1]:ref_point[1][1], ref_point[0][0]:ref_point[1][0]]
                    cv2.imwrite(output_path, crop)
                    print(f"Cropped and saved: {output_path}")
                break

            elif key == ord("q"):
                cv2.destroyAllWindows()
                return

        cv2.destroyAllWindows()

# مسیرها

```
