# Creating our Model

## y_prediction function

```python
def y_pred(W, b, x):
    y_hat = np.dot(x, W.T) + b
    return y_hat
```

```python
W = np.ones(x_train_norm.shape[1])
b = 0
y_pred(W, b, x_train_norm)
```

## Cost Function

```python
def compute_cost(W, b, x, y):
    m = x.shape[0]
    J = (y_pred(W, b, x) - y) ** 2
    j = J.sum()
    j = j / (2 * m)
    return j
    
```

```python
W = np.ones(x_train_norm.shape[1])
b = 1
compute_cost(W, b, x_train_norm, y_train_norm)
```

## Gradient Function

```python
def compute_gradient(W, b, x, y):
    m = x.shape[0]
    
    y_hat = y_pred(W, b, x)
    error = y_hat - y
    
    d_W = np.dot(error.T, x) / m
    d_b = error.sum() / m
    
    return d_W, d_b
    
```

```python
compute_gradient(W, b, x_train_norm, y_train_norm)
```

## Gradient descent Function

```python
def gradient_descent(W_0, b_0, alpha, num_iter, x, y):
    cost_history = []
    W, b = W_0, b_0
    
    for i in range(num_iter):
        d_W, d_b = compute_gradient(W, b, x, y)
        W = W - alpha * d_W
        b = b - alpha * d_b
        
        cost_history.append(compute_cost(W, b, x, y))
        if i % 100 == 0:
            print(f'cost at iteration {i} is {compute_cost(W, b, x, y)}')
            
    return W, b, cost_history
            
```

```python
W_0 = np.ones(x_train_norm.shape[1])
b_0 = 1
alpha = 0.1
num_iter = 1000

W_opt, b_opt, cost_hist = gradient_descent(W_0, b_0, alpha, num_iter, x_train_norm, y_train_norm)
```

```python
plt.plot(np.arange(num_iter), cost_hist)
plt.show()
```

```python
def box_plot(data, index):
    df = data.copy()
    Q1 = np.percentile(df.iloc[:, index], 25)
    Q3 = np.percentile(df.iloc[:, index], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    box = df[(df.iloc[:, index] >= lower_bound) & (df.iloc[:, index] <= upper_bound)]
    return box
```

```python
x_train_norm = x_train / x_train.max()
y_train_norm = y_train / y_train.max()

x_test_norm = x_test / x_test.max()
y_test_norm = y_test / y_test.max()
```

## y_prediction function

```python
def y_pred_nl(W, x):
    return np.polyval(W, x)
```

```python
W = np.ones(5)
y_pred_nl(W, x_train_norm)
```

## Cost function 

```python
def compute_cost_nl(W, x, y):
    m = len(x)
    j = (y_pred_nl(W, x) - y) ** 2
    j = j.sum() / (2 * m)
    return j
    
```

```python
W = np.ones(5)
compute_cost_nl(W, x_train_norm, y_train_norm)
```

## Gradient Function

```python
def compute_gradient_nl(W, x, y):
    m = len(x)
    n = len(W)
    
    y_hat = y_pred_nl(W, x)
    error = y_hat - y
    
    d_W = np.zeros_like(W)
    for i in range(n):
        d_W[i] = (error * (x **(n-i-1))).sum()
        
    d_W = d_W / m
    
    
    return d_W
        
```

```python
W = np.ones(5)
compute_gradient_nl(W, x_train_norm, y_train_norm)
```

## Gradient Descent

```python
def gradient_descent_nl(W_0, x, y, num_iter, alpha):
    cost_hist = []
    W = W_0
    for i in range(num_iter):
        
        d_W = compute_gradient_nl(W, x, y)
        W = W - alpha * d_W
        
        cost_hist.append(compute_cost_nl(W, x, y))
        if i % 100 == 0:
            print(f'cost at iteration {i} is {compute_cost_nl(W, x, y)}')

        
    return W, cost_hist
```
